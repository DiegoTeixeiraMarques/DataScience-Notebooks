```python
import pandas as pd
```

```python
# Carregas os dados

df = pd.read_csv('precipitações.txt')
df.columns = ['dados']
df.head()
```

```python
# Faz os splits

df1 = pd.DataFrame()
df1['part1'] = df.dados.str.split('.').str[0]
df1['part2'] = df.dados.str.split('_').str[2].str.split('.').str[0]
df1['part3'] = df.dados.str.split('_').str[-1].str.split(':').str[0]
df1['part4'] = df.dados.str.split('_').str[-1].str.split(':').str[1]
df1
```

```python
# Verifica os valores na coluna 'part3' para remover e deixar somente os meses

df1.part3.value_counts()
```

```python
# Remove os valores de 'values', mentem só os meses

values=['MUNICIPIO', 'CD', 'ANUAL', 'VE']
df2 = df1[~df1.part3.isin(values)]
df2.drop(columns=['part1', 'part2'], inplace=True)
df2
```

```python
# Pega as precipitações por mês

dados = df2.values.tolist()
dados
```

```python
# Pega os municípios

municipios = df1[df1.part3 == 'MUNICIPIO']
municipios.part4 = municipios.part4.str.split(' ').str[1]
municipios.part4 = municipios.part4.astype('int64')
#municipios.dtypes
municipios.drop(columns=['part1', 'part2'], inplace=True)
municipios_list = municipios.values.tolist()
#type(municipios_list[0][1])
municipios_list
```

```python
# Junta precipitações com municípios a cada 12 linhas percorridas

linha = 0
posicao = 0
base = []

for i in dados:
    if linha != 12:
        base.append([municipios_list[posicao][1], i[0], i[1]])
        linha = linha + 1 
    else:
        linha = 0
        posicao = posicao + 1
base
```

```python
# Carrega base do IBGE que contém o cod e nome dos municípios

df_cod_mun = pd.read_csv('municipios.csv', sep=';')
df_cod_mun.columns = ['cod', 'mun']
base_ibge = df_cod_mun.values.tolist()
base_ibge
```

```python
# Junta o nome do município a partir do cod do IBGE

for i in base:
    for j in base_ibge:
        #print('Base: ', type(i[0]), 'Munic: ', type(j[0]))
        if i[0] == j[0]:
            i.append(j[1])
            #print(j[1])
```

```python
base
```

```python
# Cria Dataset final

df_dados = pd.DataFrame(base, columns=['cod_mun', 'mes', 'pre', 'municipio'])
df_dados
```
